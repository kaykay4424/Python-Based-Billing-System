```python
print(" Welcome to M&K Innovations Pvt. Ltd, \n")
print(" Please provide us with the following information so that we may help you in the best way possible:) \n")

name = input(" Please enter your name: ")
city = input(" Please enter your city: ")

house_types = {
    1: 'Duplex',
    2: 'Apartment',
    3: 'Villa',
    4: 'Flats',
    5: 'Bungalow',
    6: 'Farmhouse',
    7: 'Penthouse'
}

print("\n Please select the type of your house:")
for code, htype in house_types.items():
    print(f"{htype} = {code}")

while True:
    try:
        h = int(input("\n Enter house type code: "))
        if h in house_types:
            break
        else:
            print(" Invalid code. Try again.")
    except ValueError:
        print(" Enter a number.")

a = int(input("\n Enter the area in sq.ft: "))

material_input = input("\n Will you provide the material? (yes/no): ").lower()
mc = 0 if material_input == "yes" else 200000

flooring_rates = {
    'hardwood': 200,
    'tile': 50,
    'marble': 150,
    'vinyl': 100,
    'carpet': 70
}

print("\n Flooring Options:")
for k, v in flooring_rates.items():
    print(f"{k.capitalize()} (Rs.{v}/sq.ft.)")

while True:
    flooring = input("\n Enter flooring type: ").lower()
    if flooring in flooring_rates:
        break
    else:
        print(" Invalid choice. Try again.")
fa = int(input(" Enter flooring area in sq.ft: "))
fc = flooring_rates[flooring] * fa

ceiling_rates = {
    'metal': 275,
    'pop': 55,
    'gypsum': 70,
    'aluminium': 200,
    'wooden': 150
}

print("\n Ceiling Options:")
for k, v in ceiling_rates.items():
    print(f"{k.capitalize()} (Rs.{v}/sq.ft.)")

while True:
    ceiling = input("\n Enter ceiling type: ").lower()
    if ceiling in ceiling_rates:
        break
    else:
        print(" Invalid choice. Try again.")
ca = int(input(" Enter ceiling area in sq.ft: "))
cc = ceiling_rates[ceiling] * ca

wall_rates = {
    'distemper': 30,
    'enamel': 25,
    'matte': 50,
    'high gloss': 60,
    'semi gloss': 40
}

print("\n Wall Finish Options:")
for k, v in wall_rates.items():
    print(f"{k.title()} (Rs.{v}/sq.ft.)")

while True:
    wall = input("\n Enter wall finish type: ").lower()
    if wall in wall_rates:
        break
    else:
        print(" Invalid choice. Try again.")
wa = int(input(" Enter wall area in sq.ft: "))
wc = wall_rates[wall] * wa

lc = a * 50  # Labour cost

# Payment Method
payment_methods = ['debit card', 'credit card', 'cheque', 'cash']
mp = input("\n Enter preferred mode of payment: ").lower()
if mp not in payment_methods:
    mp = "unspecified"

# Totals
ta = mc + lc + fc + cc + wc
tax = 0.18 * ta
gt = ta + tax

# Final Bill
print("\n" + "="*60)
print("                M&K INNOVATIONS PVT. LTD.")
print("            Innovation becomes reality here")
print("="*60)
print("Dated: 23/07/2025                    Bill No: BKMD14-284")
print(f"Customer Name: {name}")
print(f"City: {city}")
print(f"Type of House: {house_types[h]}")
print(f"Area: {a} sq.ft.")
print("-" * 60)
print(f"Material Charges:                  Rs. {mc}")
print(f"Labour Charges:                    Rs. {lc}")
print(f"Flooring Charges ({flooring}):     Rs. {fc}")
print(f"Ceiling Charges ({ceiling}):       Rs. {cc}")
print(f"Wall Finish Charges ({wall}):      Rs. {wc}")
print("-" * 60)
print(f"Total Amount:                      Rs. {ta}")
print(f"Tax (18%):                         Rs. {tax:.2f}")
print(f"Grand Total:                       Rs. {gt:.2f}")
print(f"Mode of Payment:                   {mp.capitalize()}")
print("="*60)
print("\nTHANK YOU FOR SHOPPING WITH US!")
print(f"\n{name}")
print("____________________________")
print("Customer's Signature")





import json
import os


bill_data = {
    "name": name,
    "city": city,
    "house_type": house_types[h],
    "area_sqft": a,
    "material_provided": material_input,
    "material_cost": mc,
    "labour_cost": lc,
    "flooring_type": flooring,
    "flooring_area": fa,
    "flooring_cost": fc,
    "ceiling_type": ceiling,
    "ceiling_area": ca,
    "ceiling_cost": cc,
    "wall_finish": wall,
    "wall_area": wa,
    "wall_cost": wc,
    "total_amount": ta,
    "tax": round(tax, 2),
    "grand_total": round(gt, 2),
    "payment_mode": mp
}


filename = "billing_data.json"


if os.path.exists(filename):
    with open(filename, 'r') as f:
        try:
            all_data = json.load(f)
        except json.JSONDecodeError:
            all_data = []
else:
    all_data = []


all_data.append(bill_data)

with open(filename, 'w') as f:
    json.dump(all_data, f, indent=4)

print("\n[âœ…] Billing record saved to billing_data.json")

```
